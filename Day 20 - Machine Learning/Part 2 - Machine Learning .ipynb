Things to do:
- Machine Learning
- Types:
    - Supervised Learning 
        - we have huge amounts of data
        - we train algorithms to learn patterns in LABELLED DATA
        - make decisions
    - Unsupervised Learning
        - we use huge amounts of UN-LABELLED DATA
        - make decision
    - Reinforcement Learning
        - we give feedback to our model, to teach it the favourable pattern

```python
# general steps in ML

# import
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (mean_absolute_error, 
                             mean_squared_error, 
                             root_mean_squared_error, 
                             r2_score)


# Data Acquisition - getting the data
# DATASET - inputs and labels
df = sns.load_dataset('tips')


# Feature Selection
# in order to train the model, we need to separate the input data and the target/label data
# Data Split - let's save inputs in X and outputs in y
X = df[['total_bill']]
y = df['tip']


# TRAIN TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, # testing data = 20% 
                                                    random_state=42 # used to break the data from a certain random
                                                    # like setting the seed
                                                    # to repeat the same split
                                                    )

# let's create model objects --- Algorithm
model = LinearRegression()

# Turn Algorithm ==> Model
# TRAINING process
model.fit(X_train, y_train)
# hey algo, please learn patterns in this data, and become our Trained Model

# Trained model is used to make predictions
# on data that model has never seen 
# but we do have the outputs for X_test
predictions = model.predict(X_test)

# How do we say wheater testing went well or not
# we give a score --- 
error = mean_absolute_error(y_test, predictions)
print("Error in our model =  ",error*100, "%")

newbill = [[12.00]]
tip = model.predict(newbill)
print(f"The prediction of tips for a bill = {newbill} is ----> tip = {tip}")
```

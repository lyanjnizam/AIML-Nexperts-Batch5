Precision => focus on False Alarms

- eg. bad customer experience - bringing a non smoker an ashtray because model said smokerrrr


Recall => focus on Missed Cases
- use when False Negatives are important and cannot be missed
- eg. covid case missed - setting a covid infected person free, because model said, no covid

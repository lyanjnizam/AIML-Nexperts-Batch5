- `MODULE` is a single file - like a single python script
- `PACKAGE` 
    - a folder that holds multiple related files together
    - multiple modules can exist in a package
- `LIBRARY` are packages that we can install and use 
    - Bundles **packages** and **modules** together to solve a big, specific problem
    - a library can hold multiple packages too
    - a Giant Tool box
    - we use the `pip` package of python, to install a libarary
    - example:
        - `pip` install `requests` `pillow` `numpy` `pandas`

- Library examples:
    - `requests` - used to connect to internet 
    - `numpy` - used for Numbers, numerical operations
    - `pandas`- used for data analsis, deals with tabular data as well
    - `matplotlib` - used for creating visuals like bar graph, line chart etc
    - there are libraries for every single thing you can imagine in python

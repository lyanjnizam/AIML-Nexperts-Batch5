2. Create a line plot of the sigmoid function/ Logistic function
    - y = 1/ (1 - e**-x)
    - create a sigmoid function
    - define y and x
    - then plot



```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.arange(0,11,1) # Create a list from 0-10
x = np.linspace(-10, 10, 100)

y = 10 * x + 25

def sigmoid(x):
    return 1 /(1 + np.exp(-x))
 

fig, ax = plt.subplots(1, 2, figsize = (12, 4))

ax[0].plot(x,y)
ax[0].axhline(0, color = 'red', linewidth = 0.5)
ax[0].axvline(0, color = 'red', linewidth = 0.5)
ax[0].set_title("Linear Function (function of a line)")
ax[0].grid()

ax[1].plot(x, sigmoid(x))
ax[1].axhline(0, color = 'red', linewidth = 0.5)
ax[1].axvline(0, color = 'red', linewidth = 0.5)
ax[1].axhline(0.5, color = 'red', linewidth = 0.5)

ax[1].set_title("Logistic Function (Sigmoid Curve)")
ax[1].grid()

plt.show()
```

# Linear Regression
- Is a Supervised Learning Algorithm
- In past and still as well in statistics, we used to do Regression Analysis
- We took data and used to try and understand a trend from the data
    - by calculating corelation
    - by trying to draw a line of best fit
- eg. If we had tried to test the efficacy of a drug
    - medical researcher would use linear regression to determine if
        - Taking higher dose ---> leads to lower blood pressure
- 

In Machine Learning, LR is a predictive algorithm
- LR learns from patterns in training data and can make predictions based on its training on **unseen data**
- e.g. House Price Prediction
    - you tell me the features of the house ---> algo gives you the predicted price that you can fetch
